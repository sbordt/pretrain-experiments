```python
# for all files in "../resou/train-once-answer-all/" iterate through them
from pretrain_experiments.script_utils import load_jsonl, save_jsonl

import os
directory = "../resources/train-once-answer-all/benchmark-dependence"
for filename in os.listdir(directory):
    if filename.endswith(".jsonl"):
        filepath = os.path.join(directory, filename)
        print(filepath)
        content = load_jsonl(filepath)
        benchmark_questions = [{"text": q["prompt"]} for q in content if q["label"] == q["idx"]]
        filename = filename.replace(".jsonl", "-filtered.jsonl")
        save_jsonl(benchmark_questions, os.path.join("./", filename))
```
